# Styler -- Artistic Neural Network
Make Neural Net, paint for you!


Styler is a deep learning tool, in which style of one image is applied on style of another image!

It is based on transfer learning, and it is implementation of the research paper  "[Image Style Transfer by CNN](https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf)"

Separating Style and Content
Style transfer relies on separating the content and style of an image. It takes one content image and one style image, we aim to create a new, target image which contains the desired content and style components:

In Layman's term, style inculdes color and texture, which we transfer on the content image! A content image can be assumed of the main object.

We'll use a pre-trained VGG19 Net to extract content or style features from a passed in image.

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

import torch
import torch.optim as optim
from torchvision import transforms, models
from tqdm import tqdm
%matplotlib inline
```

 In texture trans-
fer the goal is to synthesise a texture from a source image
while constraining the texture synthesis in order to preserve
the semantic content of a target image.

As per the paper, features of VGG19 net (pretrained) are used.

Texture Transfer algorithm, constraints a texture synthesis method by feature representation from state of the art CNN

```python
model = models.vgg19(pretrained=True).features
for param in model.parameters():
    param.require_grad=False
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)
```

```python
def load_image(img_path, max_size=400, shape=None):
    
    image = Image.open(img_path).convert('RGB')
    if max(image.size) > max_size:
        size = max_size
    else:
        size = max(image.size)
        
    if shape is not None:
        size = shape
        
    in_transform = transforms.Compose([
                        transforms.Resize(size),
                        transforms.ToTensor(),
                        transforms.Normalize((0.485, 0.456, 0.406), # ImageNet Standards 
                                             (0.229, 0.224, 0.225))
    ])

    image = in_transform(image)[:3,:,:].unsqueeze(0)
    
    return image
```

```python
def img_cnvrt(tensor):    
    image = tensor.to("cpu").clone().detach()
    image = image.numpy().squeeze()
    image = image.transpose(1,2,0)
    # image*std + mean = unnormalizing the image
    image = image * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))
    image = image.clip(0, 1)

    return image
```

```python
content = load_image('assets/taj_mahal.jpg').to(device)
# Resize style to match content, makes code easier
style = load_image('assets/wp3.jpg', shape=content.shape[-2:]).to(device)
```

###  Content (Left) Style (Right)

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.imshow(img_cnvrt(content))
ax2.imshow(img_cnvrt(style))
```

## Designing the algorithm

```python
def get_features(image, model, layers=None):
    
    ## We Need only these layers for the content and style representations of an image as per the paper.
    if layers is None:
        layers = {'0': 'conv1_1', '5': 'conv2_1',
                 '10': 'conv3_1', '19': 'conv4_1',
                 '21': 'conv4_2', '28': 'conv5_1'}
        
    features = {}
    x = image
    for name, layer in model._modules.items():
        x = layer(x)
        if name in layers:
            features[layers[name]] = x
            
    return features
```


Higher layers in the network
capture the high-level content in terms of objects and their
arrangement in the input image but do not constrain the ex-
act pixel values of the reconstruction very much 

In contrast, reconstructions from
the lower layers simply reproduce the exact pixel values of
the original image

**TL;DR** <br>
Higher layer => More of Content Representation (shape, outline, edges etc)
Lower Layer => More of style(color, texture etc)

**For Style** <br>
To obtain a representation of the style of an input image,
we use a feature space designed to capture texture information.
<br>
"It consists of the correlations between the different filter responses, where the
expectation is taken over the spatial extent of the feature maps."

These correlations at each layer is given by **Gram Matrix**


### Gram Matrix 

Correlations at each layer is given by Gram Matrix

* First We Vectorize the value in one layer for one layer for each layer in depth d.
* Multiply this matrix by it's transpose
  * This operation treats each value inf eature map as an individual sample unrealted in space to other values.
  
* Resultant GramMatrix contains non localised information about the layer
  * Non Localised information implies information that still be ther even if an image was shuffled around in space!
  


```python
def gram_matrix(tensor):
    batch_size, d, h, w = tensor.cpu().size()
    mat = tensor.view(d, h*w)
    gram = torch.mm(mat, mat.t())
    return gram 
```

```python
content_features = get_features(content, model)
style_features = get_features(style, model)

style_grams = {layer: gram_matrix(style_features[layer]) for layer in style_features}


target = content.clone().requires_grad_(True).to(device) # requires_grad=True, because we want to get the feature for the content image.
```

Input image x in encoded in each layer of CNN by the filter responses to that image.
So, we'll assign **weights** that will be telling our algo the amount that each layer will contribute!

<br>Weights for each style layer ; All these layers are the tested layers in the paper
As stated above, higher layers => more content and lower layers => More style
so, weighting earlier layers more will result in larger style artifacts!

```python
# excluding `conv4_2`, since it is for content representation
style_weights = {'conv1_1': 0.1,
                 'conv2_1': 0.9,
                 'conv3_1': 0.9,
                 'conv4_1': 0.6,
                 'conv5_1': 0.5}

content_weight = 1  # alpha
style_weight = 1e6  # beta
```

### Losses!
 * Content Loss
     * The mean squared difference between the target and content features at layer conv4_2
 * Style Loss
     * We'll calculate the gram matrix for the target image i.e target_gram and style image i.e style_gram at each of these layers(layers in style weight) and compare those gram matrices.
     
 * Total Loss
     * content_weight(alpha) * contentLoss + style_weight(beta)*styleLoss
 

## Training our algo, Core!

```python
# for displaying the target image, intermittently
show_every = 400

optimizer = optim.Adam([target], lr=0.03)
steps = 5000

for step in tqdm(range(1, steps+1)):
    
    target_features = get_features(target, model)
    content_loss = torch.mean((target_features['conv4_2']-content_features['conv4_2'])**2)
    
    style_loss = 0
    # iterate through each style layer and add to the style loss
    for layer in style_weights:
        target_feature = target_features[layer]
        _, d, h, w = target_feature.shape
        target_gram = gram_matrix(target_feature)
        
        style_gram = style_grams[layer]

        layer_style_loss = style_weights[layer]*torch.mean((target_gram-style_gram)**2)
        
        # adding to the style loss, because style_loss represents the overall style loss!
        style_loss += layer_style_loss / (d * h * w)
        
        
    total_loss = content_weight*content_loss + style_weight*style_loss
    
    # updating the target image
    optimizer.zero_grad()
    total_loss.backward(retain_graph=True)
    optimizer.step()
    
    # display intermediate images and print the loss
    if  step % show_every == 0:
        print('Total loss: ', total_loss.item())
        plt.imshow(img_cnvrt(target))
#         save_image(temp.permute(2,0,1).unsqueeze(0), "img_at_loss"+str(ii)+'.png')
        plt.show()
#         plt.savefig("img_at_loss%d"%ii)
```
